```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt
at=210
void_fraction= np.linspace(0.87, 0.98, 100)
Dp_list=[]

for void_fraction1 in void_fraction:
  Dp=6*(1-void_fraction1)/(at*0.3) 
  Dp_list.append(Dp)
  if 0.010>Dp>0.008:
    print(void_fraction1,  Dp  )
print(Dp_list)
# plotting the points
plt.plot(void_fraction, Dp_list)
# naming the x axis
plt.xlabel('void Fraction, X - axis')
# naming the y axis
plt.ylabel(' Dp in meter, y - axis')
 
# function to show the plot
plt.show()
```
